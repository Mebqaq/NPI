```python
import pandas as pd

NPI = pd.read_csv(r'data/decoupling/data_手动修正.csv', index_col='Unnamed: 0')
NPI.rename(columns={
    'time': 'NPI',
}, inplace=True)
NPI
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.model_selection import ParameterSampler
from mpl_toolkits.mplot3d import Axes3D
import random


features = NPI[['intensity', 'umap_1', 'umap_2']]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)


param_dist = {
    "n_clusters": list(range(2, 11))
}

best_score = -1
best_kmeans = None
random.seed(42)
n_iter = 20

for params in ParameterSampler(param_dist, n_iter=n_iter, random_state=42):
    kmeans = KMeans(n_clusters=params["n_clusters"], random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    if score > best_score:
        best_score = score
        best_kmeans = kmeans
        best_labels = labels

print(f"最佳聚类数: {best_kmeans.n_clusters}，轮廓系数: {best_score:.3f}")

NPI['cluster'] = best_labels


```

```python
from scipy.spatial.distance import cdist

cluster_centers = NPI.groupby('cluster')[['umap_2', 'umap_1']].mean()

closest_cities = []

for cluster_id, center in cluster_centers.iterrows():
    cluster_data = NPI[NPI['cluster'] == cluster_id].copy()
    
    distances = cdist(cluster_data[['umap_2', 'umap_1']], [center])
    
    min_index = distances.argmin()
    closest_city = cluster_data.iloc[min_index]
    
    closest_cities.append({
        'cluster': cluster_id,
        'city_en': closest_city['city_en'],
        'distance': distances[min_index][0]
    })

closest_df = pd.DataFrame(closest_cities)
print(closest_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects  # 【新增】仅用于给文字加白色描边，不改变字号
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from adjustText import adjust_text
from matplotlib.cm import get_cmap

# ==================== 完全保留你原本的论文基础字号 ====================
plt.rcParams.update({
    'font.family': 'Microsoft YaHei',
    'font.size': 16,
    'axes.labelsize': 14,
    'legend.fontsize': 14,
    'legend.title_fontsize': 14,
    'xtick.labelsize': 13,
    'ytick.labelsize': 13,
})

NUM_CLUSTERS = 6
# 【优化】改用 tab10 色系，亮度比 Spectral 更均衡，白底上不会有看不清的亮黄色
cmap = get_cmap('tab10')
colors = [cmap(i) for i in range(NUM_CLUSTERS)]
NPI['cluster_color'] = NPI['cluster'].apply(lambda x: colors[int(x) % NUM_CLUSTERS])

plt.figure(figsize=(12, 8))

# 【优化】精细控制气泡大小，缩放系数从 0.01 降到 0.005，防止气泡太大遮挡过多文字
scale = 0.005

sc = plt.scatter(
    NPI['umap_2'],
    NPI['umap_1'],
    c=NPI['cluster_color'],
    s=NPI['intensity'] * scale,
    alpha=0.45,             # 【优化】稍微提高一点透明度，让重叠处的地名隐约可见
    edgecolors='#444444',   # 【优化】软化纯黑边框，使其更具高级感
    linewidths=0.8,
    zorder=2,
    rasterized=True
)

# 【优化】保持原字号 10 不变，但通过 path_effects 加上白色微型描边
# 这样名字就算掉进深色气泡里，也有一圈白边保护，清晰可读
texts = []
for _, row in NPI.iterrows():
    t = plt.text(
        row['umap_2'], row['umap_1'], row['city_en'], 
        fontsize=8,        # 严格保留你原本的字号
        ha='center', va='center', color='#222222', zorder=4
    )
    # 核心黑科技：增加半透明白边，不增大字号，但可读性翻倍
    t.set_path_effects([path_effects.withStroke(linewidth=1.5, foreground="white", alpha=0.8)])
    texts.append(t)

# 【优化】配置更强的排斥逻辑，让文字自动寻找气泡稀疏的空隙
adjust_text(
    texts,
    objects=sc,                             # 【关键】让文字同时躲避气泡本身
    only_move={'points': 'xy', 'texts': 'xy'},
    arrowprops=dict(arrowstyle='-', color='#999999', lw=0.6, alpha=0.6), # 优雅的灰色指引线
    expand_text=(1.3, 1.3),
    expand_points=(1.3, 1.3),
    force_text=(0.2, 0.4),
    force_points=(0.2, 0.4)
)

# ----------------- 图例部分 -----------------
cluster_handles = [
    Line2D([0], [0], marker='o', color='w',
           markerfacecolor=colors[i], markersize=10, alpha=0.7,
           markeredgecolor='none', # 去掉生硬白边
           label=f'Cluster {i}')
    for i in range(NUM_CLUSTERS)
]

cluster_legend = plt.legend(
    handles=cluster_handles,
    title='Cluster',
    loc='lower left',
    borderpad=1,
    labelspacing=1.2,
    markerscale=1
)
plt.gca().add_artist(cluster_legend)

# 【优化】精准还原气泡真实尺寸的图例逻辑
size_legend_vals = np.linspace(NPI['intensity'].quantile(0.15), NPI['intensity'].quantile(0.85), 3)
size_handles = [
    Line2D([0], [0], marker='o', color='w', label=f'{val:.0f}',
           markerfacecolor='#888888', 
           # 【修正】Matplotlib 的 markersize 对应的是圆圈直径，等效于 sqrt(面积)
           # 必须使用 np.sqrt(val * scale) 才能在视觉上与图中的气泡大小一模一样
           markersize=np.sqrt(val * scale), 
           alpha=0.5, markeredgecolor='none')
    for val in size_legend_vals
]

plt.legend(
    handles=size_handles,
    title='Intensity',
    loc='upper right',
    borderpad=1,
    labelspacing=1.5, # 【优化】加大间距，防止大图例符号相互踩踏
    markerscale=1
)

# ----------------- 细节边框与辅助 -----------------
plt.gca().set_axisbelow(True)
plt.grid(True, linestyle='--', color='#e8e8e8', linewidth=0.8) # 增加学术淡淡的网格

plt.xlim(left=NPI['umap_2'].min() - 0.6, right=NPI['umap_2'].max() + 0.6)
plt.ylim(bottom=NPI['umap_1'].min() - 0.6, top=NPI['umap_1'].max() + 0.6)

# 软化坐标轴边框颜色，让主体更突出
for spine in plt.gca().spines.values():
    spine.set_color('#cccccc')
    spine.set_linewidth(0.8)

plt.xlabel('Umap 2')
plt.ylabel('Umap 1')
plt.tight_layout()
# plt.savefig('cluster.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# ==================== 严格保留你原有的论文基础字号 ====================
plt.rcParams.update({
    'font.family': 'Microsoft YaHei',
    'font.size': 20,
    'axes.labelsize': 22,
    'legend.fontsize': 17,
    'legend.title_fontsize': 18,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
})

df = local.copy()
features = ['LST', 'Precipitation', 'Sunset', 'GDP', 'PctOlderAdult', 'PctPrimary', 'PctSecondary', 'PctTertiary']
df_abs = df.copy()
df_abs = df_abs.merge(NPI[['city', 'cluster']], on='city', how='left')

df_long = df_abs.melt(
    id_vars='cluster',
    value_vars=features,
    var_name='feature',
    value_name='shapley'
)

fig, ax = plt.subplots(figsize=(16, 7))
unique_clusters = sorted(df_long['cluster'].unique())
n_features = len(features)
group_width = 0.8  
box_width = group_width / n_features 

# ==================== 低饱和度莫兰迪色系 ====================
colors = [
    '#7bccc4',  # LST (柔和青蓝)
    '#a8ddb5',  # Precipitation (温润淡绿)
    '#bcbddc',  # Sunset (优雅浅紫)
    '#fbb4ae',  # GDP (知性淡红)
    '#9ecae1',  # PctOlderAdult (静谧天蓝)
    '#fddaec',  # PctPrimary (内敛淡粉)
    '#fed98e',  # PctSecondary (浅砂暖黄)
    '#b3cde3'   # PctTertiary (高级雾蓝)
]
color_map = dict(zip(features, colors))

for cluster_id in unique_clusters:
    df_cluster = df_long[df_long['cluster'] == cluster_id]
    
    # 核心：每个 cluster 内部按 Shapley 绝对值均值从大到小排序
    sort_metric = df_cluster.groupby('feature')['shapley'].apply(lambda x: x.abs().mean())
    sorted_features_in_cluster = sort_metric.sort_values(ascending=False).index.tolist()
    
    for i, feature_name in enumerate(sorted_features_in_cluster):
        position = cluster_id - group_width / 2 + box_width / 2 + i * box_width
        box_data = df_cluster[df_cluster['feature'] == feature_name]['shapley']
        
        # 🎯 获取当前特征真正对应的莫兰迪色彩
        current_color = color_map[feature_name]
        
        # ==================== 无边框 + 白色中线逻辑 ====================
        bplot = ax.boxplot(
            box_data,
            positions=[position],
            widths=box_width * 0.9, 
            patch_artist=True,
            showfliers=False,
            # 将边框 edgecolor 设为和填充色完全一致，实现扁平无边框
            boxprops=dict(facecolor=current_color, edgecolor=current_color, linewidth=0),
            # 须线和触须帽也使用优雅的淡灰色，减弱视觉干扰
            whiskerprops=dict(color='#cccccc', linewidth=1.0),
            capprops=dict(color='#cccccc', linewidth=1.0),
            # 中线换成纯白色，凸显高级感
            medianprops=dict(color='#FFFFFF', linewidth=1.6) 
        )
        
        # 确保填充色与边框色完全融为一体
        for patch in bplot['boxes']:
            patch.set_facecolor(current_color)
            patch.set_edgecolor(current_color)
            
        mean_abs_value = box_data.abs().mean()
        
        # ==================== 🛠️ 【修正】让圆圈的颜色与当前箱体完美对应 ====================
        # 方案：白底 + 当前特征颜色的粗边框（这样既能看出是个圆圈，颜色又能完美呼应）
        ax.plot(position, mean_abs_value,
                marker='o',
                markersize=6.5,
                markerfacecolor='white',         # 圆圈内部填充白色，保持清爽
                markeredgecolor=current_color,   # 👈 修正：边框颜色严格对应当前箱体颜色
                markeredgewidth=1.5,             # 稍微加粗边框，让颜色更显眼
                zorder=10)

ax.set_xticks(unique_clusters)
ax.set_xticklabels([str(c) for c in unique_clusters])
ax.set_xlabel('Cluster', labelpad=12)
ax.set_ylabel('Shapley value', labelpad=12)

# 精细化背景网格线
ax.set_axisbelow(True)
ax.grid(axis='y', linestyle='--', color='#e2e2e2', linewidth=0.8)

# 软化坐标轴外边框
for spine in ax.spines.values():
    spine.set_color('#cccccc') 
    spine.set_linewidth(0.8)

# 图例色块保持去边框
legend_patches = [Patch(facecolor=color_map[feature], label=feature, edgecolor='none') for feature in features]
ax.legend(handles=legend_patches, title='Variable', ncol=2, loc='upper left', borderpad=0.8, labelspacing=0.6)

plt.tight_layout()
plt.savefig('boxplot', bbox_inches='tight', dpi=300)
plt.show()
```
